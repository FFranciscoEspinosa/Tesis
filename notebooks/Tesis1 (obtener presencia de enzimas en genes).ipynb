<h1>Función principal que crea el data frame, de acuerdo a el filo y los archivos donde se encuentra la información</h1>


```python
def crear_df_precencia(filo,central,carpeta):
    enz_y_comp=enzimas_nombre_y_composicion(filo,central) 
    enz,comp=list(zip(*enz_y_comp)) #Se obtienen las enzimas y su composición en dos arreglos con la función enzimas_nombre
    directorio = '/files/francisco_esp/'+filo+'/'+carpeta+'/' #Liga de la carpeta donde están los genomas
    archivos = np.asanyarray(os.listdir(directorio))
    df = pd.DataFrame(columns=enz) #Se crea un data frame con columnas los nombres de las enzimas
    for i in archivos: 
        if i.endswith(".txt"): #Se recorren los nombres de todos los achivos de la carpeta y solo se revisan los .txt
            nombre_archivo=directorio+i 
            archivo=open(nombre_archivo, "r",encoding='cp1252')
            info_archivo=archivo.read()
            archivo.close() #Se lee el archivo en el que va el ciclo
            for j in range(len(enz)): #agrega una fila e información a la posición i,j
                df.loc[i.replace('.txt',''),enz[j]]=info_archivo.count(comp[j])
    return df
```

<h3>Función que busca el nombre de las enzimas y sus composiciones</h3>

```python
def enzimas_nombre_y_composicion(filo,central):
    nombre_archivo="/files/francisco_esp/"+filo+"/"+central
    file = open(nombre_archivo, "r")
    enz_y_comp = file.read().split('>')[1:]
    enz_y_comp=depurar_enz(enz_y_comp)
    file.close()
    return enz_y_comp
```

<h3>Función que separa los datos de acuerdo a las enzimas y depura el texto</h3>

```python
def depurar_enz(enz_y_comp):
    for i in range(len(enz_y_comp)):
        enz_y_comp[i]=enz_y_comp[i].strip('\n\t').split('\n',maxsplit=1)
        enz_y_comp[i][1]=enz_y_comp[i][1].replace('\n','')
    return enz_y_comp
```
