```python
def retroprop(pesos,x,y,n,alpha=0.05):
    x1,x2=x
    for i in range(n):
        #se desempacan los pesos
        w1,w2,w3,w4,w5,w6=pesos
        #los primeros valores que reciben las neuronas son
        z1=x1*w1+x2*w3
        z2=x1*w2+x2*w4

        #La neurona final devuelve
        evaluacion=z1*w5+z2*w6
        #se tiene error
        error=evaluacion-y
        
        #Se sacan las derivadas
        derivadas=np.asanyarray([w5*x1, w6*x1, w5*x2, w6*x2,z1,z2])*(error)
        #Se calculan las derivadas por medio de desenso al gradiente
        nuevos_pesos=np.asanyarray(pesos)-alpha*derivadas

        #Actualizan los pesos
        pesos=nuevos_pesos
        print('Error anterior:', error)
        print('Nuevos pesos en la iteración',i+1,':','w_1=',pesos[0],'w_2=',pesos[1],'w_3=',pesos[2],
              'w_4=',pesos[3],'w_5=',pesos[4],'w_6=',pesos[5] )
```

```python
pesos=[0.11,0.21,0.12,0.08,0.14,0.15]
x=(2,3)
y=1

retroprop(pesos,x,y,1000)
```
